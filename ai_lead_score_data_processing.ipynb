```python
import pandas as pd
```

```python
df=pd.read_csv('enriched_with_linkedin.csv')
```

```python
df.head()
```

```python
df["created_at"] = pd.to_datetime(df["created_at"], utc=True).dt.tz_localize(None)
```

```python
from datetime import datetime
today = pd.Timestamp(datetime.utcnow())
df["domain_age"] = ((today - df["created_at"]).dt.days / 365).round(2)
```

```python
df['hq'].unique()
```

```python
import ast

def parse_hq(hq_string):
    try:
        return ast.literal_eval(hq_string) if pd.notnull(hq_string) else {}
    except Exception:
        return {}
```

```python
df["hq_dict"] = df["hq"].apply(parse_hq)
df["hq_country"] = df["hq_dict"].apply(lambda x: x.get("country", ""))
df["hq_state"] = df["hq_dict"].apply(lambda x: x.get("state", ""))
df["hq_city"] = df["hq_dict"].apply(lambda x: x.get("city", ""))
```

```python
df["industry"] = df["industry"].fillna("Unknown")
```

```python
df['hq_country'] = df['hq_country'].replace('', 'remote')
```

```python
df["linkedin_followers"] = df["linkedin_followers"].fillna(0)
```

```python
df["employee_count"] = df["employee_count"].fillna(df["employee_count"].median())
```

```python
features = ["votes_count", "reviews_count", "linkedin_followers", "employee_count", "domain_age"]
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scaled = scaler.fit_transform(df[features])
```

```python
df['email_valid']=df["email_valid"].astype(int)
```

```python
df['linkedin_enriched']=df["linkedin_enriched"].astype(int)
```

```python
# Frequency encoding
industry_freq = df["industry"].value_counts(normalize=True).to_dict()
df["industry_freq"] = df["industry"].map(industry_freq)
```

```python
hq_country_freq = df["hq_country"].value_counts(normalize=True).to_dict()
df["hq_country_freq"] = df["hq_country"].map(hq_country_freq)
```

```python
df['industry_freq'] = df['industry_freq'].fillna(0)
df['hq_country_freq'] = df['hq_country_freq'].fillna(0)
```

```python
df_scaled = pd.DataFrame(df_scaled, columns=features)
```

```python
# Combine scaled numerical + encoded categorical
final_features = pd.concat([df[['industry_freq', 'hq_country_freq','company_name','email_valid','linkedin_enriched']].reset_index(drop=True),df_scaled ], axis=1)
```

```python
final_features.head()
```

```python
final_features.info()
```

```python
final_features['founded_recently'] = (final_features['domain_age'] < 2).astype(int)
```

```python
final_features['lead_score'] = (
    0.2 * final_features['votes_count'] +
    0.1 * final_features['reviews_count'] +
    0.2 * final_features['linkedin_followers'] +
    0.1 * final_features['employee_count'] +
    0.05 * final_features['domain_age'] +
    0.1 * final_features['email_valid'] +
    0.05 * final_features['linkedin_enriched'] +
    0.05 * (1 - final_features['founded_recently']) +
    0.1 * final_features['industry_freq'] +
    0.05 * final_features['hq_country_freq']
)

# Optional normalization
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
final_features['lead_score_normalized'] = scaler.fit_transform(final_features[['lead_score']])

```
